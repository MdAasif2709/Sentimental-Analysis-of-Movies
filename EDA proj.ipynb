# Distributions

```python
col=['runtime_minutes','movie_averageRating','movie_numerOfVotes','approval_Index','budget','D_Gross','W_gross']
```

## Normal Distribution

```python
import scipy.stats as stats

alpha=0.05
normality_results=[]
for column in col:

        # Perform the Shapiro-Wilk test
         statistic, p_value = stats.shapiro(movies[column]) 
         is_normal = p_value > alpha
         if is_normal:
                normality_results.append(column) #prints columns with normal distribution
         else:
           print(column,"is not normal")
  

for i  in normality_results:
    print(i)
```

## Logarithmic Distribution

```python
import scipy.stats as stats

alpha=0.05
normality_results=[]
for column in col:
        fit = stats.lognorm.fit(movies[column])
        ks_statistic, p_value = stats.kstest(movies[column], 'lognorm', args=fit)
        is_log= p_value > alpha
        if is_log:
                normality_results.append(column)
        else:
            print(column,"is not logarithmic")
for i  in normality_results:
    print(i)
```

## Exponential Distrubution

```python
import scipy.stats as stats

alpha=0.05
normality_results=[]
for column in col:
        # Perform the Shapiro-Wilk test
         fit = stats.expon.fit(movies[column])
         ks_statistic, p_value = stats.kstest(movies[column], 'expon', args=fit, alternative='greater')
         is_expo = p_value > alpha
         if is_expo:
                normality_results.append(column)
         else:
            print(column,"is not exponential distribution")
for i  in normality_results:
    print(i)
```

# 2. F-test

## Hypothesis: There is a significant difference in movie average ratings and production budgets among different genres.

```python
f_statistic, p_value = stats.f_oneway(movies[movies['genre'] == 'Action']['movie_averageRating'],
                                      movies[movies['genre'] == 'Comedy']['movie_averageRating'],
                                      movies[movies['genre'] == 'Drama']['movie_averageRating'])

# Print the results
print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Check for statistical significance
alpha = 0.05  # Set your significance level
if p_value < alpha:
    print("There is a significant difference in movie average ratings between genres.")
else:
    print("There is no significant difference in movie average ratings between genres.")
```

# 3.Chi-square Test

## Hypothesis: There is a significant association between genre and earning

```python
import pandas as pd
import scipy.stats as stats

# Assuming you have a DataFrame named 'df' with your movie dataset
# Select the categorical variables to analyze

categorical_variable1 = 'genre'
categorical_variable2 = 'Earning'

# Create a contingency table
contingency_table = pd.crosstab(movies[categorical_variable1], movies[categorical_variable2])

# Perform the chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Print the results
print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)

# Set your significance level (alpha)
alpha = 0.05

# Check for statistical significance
if p < alpha:
    print("Reject the null hypothesis: There is a significant association between", categorical_variable1, "and", categorical_variable2)
else:
    print("Fail to reject the null hypothesis: There is no significant association between", categorical_variable1, "and", categorical_variable2)
```
